**First, let’s remove redundant data from the ‘Name’ column.**

**This involves removing punctuation, numbers, special characters, etc., and retaining only alphabets**

```python
df['name'][10]
```

```python
def clean_resturant_name(name):
    if isinstance(name,str):
        if(re.findall(r'[a-zA-Z0-9]',name)):# findall() returns a list containing all matches.
            list_name = (re.findall(r'[a-zA-Z0-9]',name))
            str_name = "".join(list_name)# join() method takes all items in an iterable and join them into one string 
            return str_name
        else:
            return name
```

**Let us now look at the ‘Rate’ column**

**We can see that we have ‘nan,’ ‘NEW,’ and ‘-‘ values that do not have any ratings, and also, the values are strings containing ‘/5’. Let us remove all insignificant data and convert the ratings into numeric values.**

**removing 'nan', 'NEW', '-' values**

**Removing '/5' from rate column**

```python
def clean_rate(rates):
    if rates =='-' or rates =='NEW':
        return np.nan
    else:
        rates_converted = float(str(rates).split('/')[0])
        return rates_converted
```

```python
df['rate']=df['rate'].apply(clean_rate)
```

# **Step 6: Exploratory Data Analysis**

**Now that we have cleaned our data, it is ready for analysis**

**1. Which are the top ten resturants according to highest votes**

```python
high_vote = df.groupby('name').agg({'votes':'max'}).sort_values(by='votes',ascending=False)[:10]
```

```python
high_name=high_vote.index
high_votes=high_vote['votes']
fig=plt.figure(figsize=(10,8))
sns.set_style(style="whitegrid")
sns.barplot(x=high_name,y=high_votes,data=high_vote)
plt.xticks(rotation=40,horizontalalignment='right',fontsize='x-large')
plt.title("Top 10 resturant with highest voting")
plt.tight_layout()
plt.show()
```

**These restaurants are the ones that receive the most votes from patrons.**

**2. which are top 10 resturant with highest rating**

```python
names=high_rate.index
ratings=high_rate['rate']
plt.figure(figsize=(10,8))
sns.set_style(style="whitegrid")
sns.barplot(x=names,y=ratings,data=high_rate,alpha=0.9)
plt.xticks(rotation=40,horizontalalignment="right",fontsize='x-large')
plt.title("Top 10 resturant with highest rating!!")
plt.tight_layout()
plt.show()
```
